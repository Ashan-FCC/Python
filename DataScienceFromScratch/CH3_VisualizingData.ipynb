```python
#Matplotlib
from matplotlib import pyplot as plt;

years = [1950,1960, 1970, 1980, 1990, 2000, 2010];
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3];

#Create a line chart. X - years, Y - GDP

plt.plot(years,gdp,color='green',marker='o',linestyle='solid');

#Add a title
plt.title('Nominal GDP');

#Add a label to the y-axis
plt.ylabel("Billions of $");

#Add a label to the x-axis
plt.xlabel("Years");

plt.show();
```

```python
#Bar Charts
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"];
num_oscars = [5,11,3,8,10];

#bars are by default of width 0.8, we add 0.1 to the left
#coordinates so that each bar is centered.
xs = [i + 0.1 for i, _ in enumerate(movies)]
print(xs);
#plot bars with left x-coordinates [xs] , heights [num_oscars]
plt.bar(xs, num_oscars);

plt.ylabel("# of Academy Awards");
plt.title("My favourite movies");

#label x-axis of with movie names at bar centers
plt.xticks([i + 0.5 for i,_ in enumerate(movies)], movies);

plt.show();
```

```python
#Barchart for a histogram
from collections import Counter
    
grades = [83, 95, 91, 87, 70, 0 ,85, 82,100,67,73,77,0 ];
#Get floor of divide
decile = lambda grade : (grade // 10) * 10
histogram = Counter(decile(grade) for grade in grades);
print(histogram);
plt.bar([x-4 for x in histogram.keys()], #shift each bar to the left by 4
       histogram.values(), #give each bar its correct height
       8); #Give each a bar a width of 8

plt.axis([-5, 105, 0 ,5]) #X axis from -5 to 105
                          #Y axis from 0 to 5
plt.xticks([10 * i for i in range(11)]); #X ranges 0, 10, ..100

plt.xlabel("Decile");
plt.ylabel("No of Students");

plt.show();
```

```python
#Line Charts
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256];
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1];

total_error = [x + y for x,y in zip(variance, bias_squared)];

xs = [i for i,_ in enumerate(variance)];

plt.plot(xs, variance, 'g-', label='variance'); #solid
plt.plot(xs, bias_squared, 'r-.', label='bias^2'); #dot-dash
plt.plot(xs, total_error, 'b:', label='total error'); #dotted

plt.legend(loc=9) # 9 means top center
plt.xlabel('model complexity');
plt.ylabel('The bias-variance tradeoff');
plt.show();
```

```python
#Scatter Plots
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends,minutes);

#label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, 
                xy=(friend_count,minute_count),#Put the label
                xytext=(5, -5),    #with point but slight offset
                textcoords='offset points'
                );
plt.title("Daily Minutes vs Number of Friends");
plt.xlabel("# of Freinds");
plt.ylabel("Daily minutes spent on site");
plt.show();


```
