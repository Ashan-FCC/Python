```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
link = "http://www.rssweather.com/wx/th/hua+hin/wx.php"

with urllib.request.urlopen(link) as url:
    html_doc = url.read()

soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
print(soup.prettify())
```
